```python
df=pd.read_csv('AWCustomers.csv')
df.head()
```

```python
df2=pd.read_csv('AWSales.csv')
df2.head()
```

```python
df3=pd.merge(df,df2,on='CustomerID',how='inner')
df3.head()
```

```python
#Feature selection

fin_df=pd.DataFrame()
fin_df['PostalCode']=df3['PostalCode']
fin_df['Age']=2024-(df3['BirthDate'].str.split('-', expand=True)[0].astype(int))
fin_df['YearlyIncome']=df3['YearlyIncome']
fin_df['Occupation']=df3['Occupation']
fin_df['Gender']=df3['Gender']
fin_df['Education']=df3['Education']
fin_df['MaritalStatus']=df3['MaritalStatus']
fin_df['NumberChildrenAtHome']=df3['NumberChildrenAtHome']
fin_df['NumberCarsOwned']=df3['NumberCarsOwned']
fin_df['AvgMonthSpend']=df3['AvgMonthSpend']
fin_df['BikeBuyer']=df3['BikeBuyer']

fin_df.head()
```

```python
fin_df.to_csv('BikeBuyer_Data.csv',index=False)
```

```python
dataFrame = pd.read_csv("BikeBuyer_Data.csv")
```

```python
#checking for null values

res=dataFrame.isnull().sum()
print(res)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
dataFrame['YearlyIncome']=scaler.fit_transform(dataFrame['YearlyIncome'].values.reshape(-1,1)).flatten()
dataFrame['AvgMonthSpend']=scaler.fit_transform(dataFrame['AvgMonthSpend'].values.reshape(-1,1)).flatten()
dataFrame['Age']=scaler.fit_transform(dataFrame['Age'].values.reshape(-1,1)).flatten()
```

```python
dataFrame.head()
```

```python
final_df=pd.get_dummies(dataFrame,columns=['Gender', 'Education', 'MaritalStatus','Occupation','PostalCode'])
final_df.replace({False: 0, True: 1}, inplace=True)
final_df.head()
```

```python
final_df.to_csv('BikeBuyer_Data_final.csv',index=False)
```
